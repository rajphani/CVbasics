**IMAGE HISTOGRAM**

```
def Histogram(D2,D1,x,y):
  for i in range(y):
    for j in range(x):
        D1[D2[j,i]] = D1[D2[j,i]]+1
  return D1
```

```
import numpy as np
preimg=cv2.imread('lena.jpg',0)
D1 = np.zeros((256,),dtype=np.float16)
x,y=img.shape
hist = Histogram(preimg,D1,x,y)
```

```
plt.plot(hist)
```

```
colors = ('b','g','r')
for i,color in enumerate(colors):
    histr = cv2.calcHist(preimg,[i],None,[256],[0,256])
    plt.plot(histr,color = color)
plt.show()
```

 **HISTOGRAM EQUALIZATION**



```
a = np.zeros((256,),dtype=np.float16)
b = np.zeros((256,),dtype=np.float16)
x,y=img.shape

for i in range(y):
    for j in range(x):
        a[img[j,i]] = a[img[j,i]]+1
```

```
temp = 1.0/(x*y)
b = np.zeros((256,),dtype=np.float16)

for i in range(256):
    for j in range(i+1):
        b[i] += a[j] * temp
    b[i] = round(b[i] * 255)
```
